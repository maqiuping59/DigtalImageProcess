```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("panda.jpg")
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title("original img")
```

```python
img.shape
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img)
plt.title("gray img")
gray_img.shape
```

```python
import numpy as np
height = img.shape[0]
width = img.shape[1]
gray_mat = np.empty((height,width),dtype = np.uint8)
for i in range(height):
    for j in range(width):
        gray_mat[i,j] = np.uint8(0.299*img[i,j,0]+0.587*img[i,j,1]+0.114*img[i,j,2])

plt.imshow(gray_mat)
cv2.imwrite("gray_mat.jpg",gray_mat)
            
```
